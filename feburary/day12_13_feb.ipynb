```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\datasets2\company_sales_data.csv")
df
```

```python
plt.title("Company profit per Month")
plt.xlabel("Month Number")
plt.ylabel("Profit in dollar")

plt.plot(df['month_number'],df['total_profit'],color='skyblue',linewidth=2)
plt.yticks(range(100000,600000,100000))
plt.xticks(range(1,13))
plt.show()
```

```python
# The line style can be written in a shorter syntax:

# linestyle can be written as ls.

# dotted can be written as :.

# dashed can be written as --.
```

```python
plt.title("Comapany sales data of last year")
plt.xlabel("Month Number")
plt.ylabel("Sold units number")

plt.yticks(range(100000,600000,100000))
plt.xticks(range(1,13))
plt.plot(df['month_number'],df['total_profit'],color='red',linestyle='--',marker='o',markerfacecolor='darkred',linewidth=3,markersize=7)
plt.legend(['Profit data of last year'],loc='lower right')
plt.show()
```

```python
plt.ylabel("Sales unit in number")
plt.xlabel("Month Number")
plt.title("Sales data")

plt.yticks(range(1000,19000,1000))
plt.xticks(range(1,13))
plt.plot(df['month_number'],df['facecream'],color='blue',linestyle='-',marker='o',linewidth=3,markersize=7,label="face cream sales data")
plt.plot(df['month_number'],df['facewash'],color='red',linestyle='-',marker='o',linewidth=3,markersize=7,label="face wash sales data")
plt.plot(df['month_number'],df['toothpaste'],color='green',linestyle='-',marker='o',linewidth=3,markersize=7,label="toothpaste sales data")
plt.plot(df['month_number'],df['bathingsoap'],color='orange',linestyle='-',marker='o',linewidth=3,markersize=7,label="bathingsoap sales data")
plt.plot(df['month_number'],df['shampoo'],color='purple',linestyle='-',marker='o',linewidth=3,markersize=7,label="shampoo sales data")
plt.plot(df['month_number'],df['moisturizer'],color='brown',linestyle='-',marker='o',linewidth=3,markersize=7,label="moisturizer sales data")

plt.legend(loc='upper left')
plt.show()
```

```python
plt.xlabel("Month number")
plt.ylabel("Number of units sold")
plt.title("Tooth paste Sales data")

plt.xticks(range(1,13))
plt.scatter(df['month_number'],df['toothpaste'],marker='o',color='blue')
plt.legend(["Tooth paste Sales data"],loc='upper left')
plt.grid(ls='--')
```

```python
plt.xlabel("Month Number")
plt.ylabel("Sales unit in Number")
plt.title("Facecream and facewash sales data")

plt.bar(df['month_number'],df['facewash'],color='blue',width=0.2,label="face cream sales data")
plt.bar(df['month_number']+0.2,df['facecream'],color='orange',width=0.2,label="face wash sales data")

l=np.arange(1,13)


plt.yticks(range(0,4500,500))
plt.xticks(l+0.1,l)   # input format is position of xticks and label of xticks

plt.grid(ls='--')
plt.legend(loc="upper left")
plt.show()
```

```python
plt.xlabel("Month Number")
plt.ylabel("Sales unit in Number")
plt.title("BathingSoap sales data")

plt.bar(df['month_number'],df['bathingsoap'],color='blue')
plt.grid(ls='--')


plt.xticks(l)
plt.show()
```

```python
plt.title("Profit data")
plt.xlabel("Profit range in dollar")
plt.ylabel("Actual profit in dollar")
plt.hist(df['total_profit'],bins=8,range=[df['total_profit'].min(),df['total_profit'].max()])
```

```python
x1=[]

x1.append((df['bathingsoap'].sum()/df['total_units'].sum())*100)
x1.append((df['facecream'].sum()/df['total_units'].sum())*100)
x1.append((df['facewash'].sum()/df['total_units'].sum())*100)
x1.append((df['moisturizer'].sum()/df['total_units'].sum())*100)
x1.append((df['shampoo'].sum()/df['total_units'].sum())*100)
x1.append((df['toothpaste'].sum()/df['total_units'].sum())*100)

print(x1)
label1=['bathingsoap','facecream','facewash','moisturizer','shampoo','toothpaste']
colors=['red','blue','orange','brown','purple','green']
plt.pie(x1,labels=label1,autopct='%.1f%%',colors=colors)
plt.legend()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.title("Sales data of bathing soap")
plt.plot(df['month_number'],df['bathingsoap'],color='black',marker='o')




plt.subplot(2,1,2)
plt.title("Sales data of facewash")
plt.plot(df['month_number'],df['facewash'],color='red',marker='o')


plt.xlabel("Month Number")
plt.ylabel("sales units in number")
plt.tight_layout()
plt.show()
```

```python
plt.title("All product sales data using stack plot")
plt.ylabel("Sales unit in number")
plt.xlabel("Month Number")
plt.xticks(range(1,13))

plt.plot([],[],color='purple',label='facecream')
plt.plot([],[],color='skyblue',label='facewash')
plt.plot([],[],color='red',label='toothpaste')
plt.plot([],[],color='black',label='bathingsoap')
plt.plot([],[],color='green',label='shampoo')
plt.plot([],[],color='olive',label='moisturizer')


colors1=['purple','skyblue','red','black','green','olive']
plt.stackplot(df['month_number'],df['facecream'],df['facewash'],df['toothpaste'],df['bathingsoap'],df['shampoo'],df['moisturizer'],colors=colors1)
plt.legend(loc='upper left')

plt.show()
```
