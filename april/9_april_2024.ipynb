```python
# code for feature selection
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as rf


x1=dataframe_disease_train.drop('prognosis',axis=1)
y1=dataframe_disease_train['prognosis']


# using select from model approach for feature selection
select=SelectFromModel(rf(n_estimators=100,random_state=0))

select.fit(x1,y1)

support=select.get_support()

# extracting important features obtained from random forest model
features=x1.loc[:,support].columns.tolist()

print(f"Important features acc to random forest: {features}")
print(f"No of features to consider after feature slecetion: {len(features)}")
print(rf(n_estimators=100,random_state=0).fit(x1,y1).feature_importances_)
```

```python
# making a datafrom for features, extracted after feature selection
dataframe_disease_after_feature_selection=dataframe_disease_train[features]
dataframe_disease_after_feature_selection
```

```python
dataframe_disease_train['prognosis'].value_counts()
```

```python
# pie chart for categories for of target variable
import matplotlib.pyplot as plt

disease_counts = dataframe_disease_train['prognosis'].value_counts()

# Extract values and indexes from the counts
values = list(disease_counts.values)
indexes = list(disease_counts.index)

plt.figure(figsize=(15,15))

plt.pie(values, labels=indexes, autopct='%1.1f%%')

# Add a title
plt.title('Distribution of Different Diseases')

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
# transforming the target label to numerical labels
encoder=LabelEncoder()
y1=encoder.fit_transform(y1)
y1
```

```python
# different model for using different algos like support vector machine,naive bayes,random forest
model_random_forest=RandomForestClassifier(n_estimators=50,max_depth=4)
model_support_vector_machine=SVC(kernel='rbf')
model_naive_bayes=GaussianNB()
```

```python
#pd.set_option('display.max_columns',None)
len(dataframe_disease_after_feature_selection.columns)
```

```python
# fitting the train data in the model
model_random_forest.fit(dataframe_disease_after_feature_selection,y1)
```

```python
# fitting the train data in the model
model_support_vector_machine.fit(dataframe_disease_after_feature_selection,y1)
```

```python
# fitting the train data in the model
model_naive_bayes.fit(dataframe_disease_after_feature_selection,y1)
```

```python
# converting the test dataset to a dataframe
dataframe_disease_test=pd.read_csv("D:\datasets2\disease_Testing.csv")
dataframe_disease_test
```

```python
# separating the independent features and target variable for test dataset
dataframe_disease_test_after_feature_selection=dataframe_disease_test[features]

y2=dataframe_disease_test['prognosis']

# transforming the target label to numerical labels
y2=encoder.fit_transform(y2)
```

```python
predictions_random_forest=model_random_forest.predict(dataframe_disease_test_after_feature_selection)
print("Accuracy of the random forest model: ",accuracy_score(y2,predictions_random_forest))
print("classification report of random forest model: \n",classification_report(y2,predictions_random_forest))
```

```python
print("confusion matric of random forest model: ")

plt.figure(figsize=(15,15))

# using heatmap for confusion matrix of random forest model
sns.heatmap(confusion_matrix(y2,predictions_random_forest),cmap='summer',annot=True)
```

```python
predictions_naive_bayes=model_naive_bayes.predict(dataframe_disease_test_after_feature_selection)

print("Accuracy of the rnaive bayes model: ",accuracy_score(y2,predictions_naive_bayes))

print("classification report of rnaive bayes model: \n",classification_report(y2,predictions_naive_bayes))
```

```python
print("confusion matric of naive bayes model: ")

plt.figure(figsize=(15,15))

# using heatmap for confusion matrix of naive bayes model
sns.heatmap(confusion_matrix(y2,predictions_naive_bayes),cmap='summer',annot=True)
```

```python
predictions_support_vector_machine=model_support_vector_machine.predict(dataframe_disease_test_after_feature_selection)

print("Accuracy of the support vector machine model: ",accuracy_score(y2,predictions_support_vector_machine))

print("classification report of support vector machine model: ",classification_report(y2,predictions_support_vector_machine))
```

```python
print("confusion matric of support vector machine model: ")

plt.figure(figsize=(15,15))

# using heatmap for confusion matrix of support vector machine
sns.heatmap(confusion_matrix(y2,predictions_support_vector_machine),cmap='summer',annot=True)
```
