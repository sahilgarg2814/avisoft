```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report
```

```python
X=dataframe.iloc[:,1:]
y=dataframe.iloc[:,0]
```

```python
from imblearn.over_sampling import SMOTE
over_sampler=SMOTE()
x_res,y_res=over_sampler.fit_resample(X,y)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_res,y_res,test_size=0.8,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model1=LogisticRegression(solver='saga',max_iter=500)
```

```python
model2=DecisionTreeClassifier()
```

```python
model3=RandomForestClassifier()
```

```python
model1.fit(xtrain,ytrain)
```

```python
predictions=model1.predict(xtest)
print(accuracy_score(ytest,predictions))
print(classification_report(ytest,predictions))
```

```python
model2.fit(xtrain,ytrain)
```

```python
predictions=model2.predict(xtest)
print(accuracy_score(ytest,predictions))
print(classification_report(ytest,predictions))
```

```python
model3.fit(xtrain,ytrain)
```

```python
predictions=model3.predict(xtest)
print(accuracy_score(ytest,predictions))
print(classification_report(ytest,predictions))
```
