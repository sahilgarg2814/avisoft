```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
df=pd.read_csv(r"D:\datasets2\homeprices.csv")
```

```python
print(df)
```

```python
plt.scatter(df['area'],df['price'],color='black')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
x=df[['area']]
y=df['price']
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
import numpy as np
```

```python
new_value=[1000,1500,2300,3540,4120,4560,5490,3460,4750,2300,9000,8600,7100
]
predictions=model.predict(pd.DataFrame(new_value))
```

```python
predictions_df=pd.DataFrame({'area':new_value,'predictions':predictions})
predictions_df.to_csv('prediction_house_price.csv')
```

```python
df2=pd.read_csv(r"D:\datasets2\canada_per_capita_income.csv")
```

```python
print(df2)
```

```python
plt.scatter(df2['year'],df2['per capita income (US$)'])
plt.xlabel('year')
plt.ylabel('per capita income')
plt.show()
```

```python
model2=linear_model.LinearRegression()
model2.fit(df2[['year']],df2['per capita income (US$)'])
```

```python
print(model2.predict([[2020]]))
```

```python
import matplotlib.pyplot as plt
plt.plot(df2[['year']],model2.predict(df2[['year']]),color='red',label='predicted per capita income')
plt.plot(2020,model2.predict([[2020]]),color='green',marker='+',label='predicted per capita income for 2020')
plt.xlabel('year')
plt.ylabel('canda per capita income')
plt.legend()
plt.show()
```
