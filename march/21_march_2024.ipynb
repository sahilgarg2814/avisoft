```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import MinMaxScaler
```

```python
df1=pd.read_csv(r"D:\datasets2\50_Startups.csv")
df1
```

```python
df2=pd.get_dummies(data=df1,prefix=['State'],drop_first=True,dtype=int)
df2
```

```python
df3=df2.drop(['Profit'],axis=1)
df3
```

```python
min_max=MinMaxScaler()
df3[['R&D Spend','Administration','Marketing Spend']]=min_max.fit_transform(df3[['R&D Spend','Administration','Marketing Spend']])
```

```python
df3
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df3,df1['Profit'],test_size=0.3,random_state=101)
```

```python
model1=LinearRegression()
model1.fit(xtrain,ytrain)
```

```python
predictions=model1.predict(xtest)
predictions
```

```python
print("models accuracy: ",r2_score(ytest,predictions))
```

```python
X=df3
print(X)
```

```python
y=df1['Profit']
```

```python
def backward_elimination(X,y,sl=0.05):
    num_features=X.shape[1]
    for i in range(0,num_features):

        Xtemp=sm.add_constant(X)
        regressor_ols=sm.OLS(y,Xtemp).fit()
        print(regressor_ols.pvalues)
        print(type(regressor_ols.pvalues))

        max_p_val=max(regressor_ols.pvalues)
        print(max_p_val)

        if max_p_val>sl:
            index_max_p_val=np.where(regressor_ols.pvalues==max_p_val)[0][0]
            print(index_max_p_val)

        X=X.drop(X.columns[index_max_p_val-1],axis=1)

        if len(X.columns)==1:
            return X
        
    return X
```

```python
print(backward_elimination(X,y))
```
